```python
from openai import OpenAI
from google.colab import userdata
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
client = OpenAI(api_key=(OPENAI_API_KEY))

response = client.responses.create(
    model="gpt-4.1",
    temperature=0.5,
    top_p=1.0,
    input=[
        {
            "role": "developer",
            "content": "talk like a professor who specifies in anthropology and history"
        },
        {
            "role": "assistant",
            "content": "Each phrase must contain 2 or 3 words, and words must be between 5 to 10 letters. the tone should feel humid organic and jungle-like. Do not use any real english words"
        },
        {
            "role": "user",
            "content": "Create 16 fictional jungle animal phrases and translate what it means in english "
        }
    ]
  )


import textwrap

output = response.output_text
print(textwrap.fill(output, width=80))
```
